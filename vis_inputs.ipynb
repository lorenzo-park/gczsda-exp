```python
import matplotlib.pyplot as plt
import numpy as np
from dataset.digit import CENDataset
import torchvision
import torch
from torchvision.transforms import transforms
```

```python
dataset = torchvision.datasets.FashionMNIST("/root/dezsda/data/FashionMNIST_G", transform=transforms.ToTensor())
plt.rcParams["figure.figsize"] = (12,12)
rotate = False
imgs = []
for i in range(4):
    img, _ = dataset[i]
    if rotate:
        imgs.append(torch.rot90(img.unsqueeze(0), k=1, dims=[-2,-1]))
    else:
        imgs.append(img.unsqueeze(0))
plt.imshow(torchvision.utils.make_grid(torch.cat(imgs, dim=0), dim=0, nrow=2).permute(1,2,0))
plt.show()
```
